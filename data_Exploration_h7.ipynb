```python
from pandas.api.types import is_numeric_dtype
```

```python
for col in data.columns:
  if is_numeric_dtype(data[col]):
    print('%s:' %(col))
    print('\t Mean = %.2f'% data[col].mean())
    print('\t Standard Deviation = %.2f' %data[col].std())
    print('\t Minimum Value = %.2f' %data[col].min())
    print('\t Maximum Value = %.2f' %data[col].max())
```

Researchers in climate science, environmental studies, or related fields may also choose to explore the frequency of climate-related disasters to contribute to the scientific understanding of climate change impacts. This research can lead to more accurate modeling and predictions.
